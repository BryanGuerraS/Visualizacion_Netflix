<h2>3. Limpieza de datos</h2>

Reconocemos que las listas 1 y 5 solo contienen películas,mientras que las listas 3 y 4 solo contienen series.<br>
La lista 2 presenta películas y series, así que vamos a separarlos.

```python
lista_2_PyS = df_netflix.Title.str.split(pat = ':', expand=True)
lista_2_PyS = lista_2_PyS[lista_2_PyS[2].isna() & lista_2_PyS[3].isna() & 
                          lista_2_PyS[4].isna() & (lista_2_PyS[1].isna() == False)]
pd.set_option('display.max_rows', 200)
lista_2_PyS.head(200)
```

Encontramos que las series contienen la palabra "Capítulo", "Episodio" o "Episode". Además, quitamos las series que vemos a simple vista.

```python
lista_2_peliculas = lista_2_PyS[(lista_2_PyS[1].str.contains('Capítulo') == False) & 
                                (lista_2_PyS[1].str.contains('Episodio') == False) &
                                (lista_2_PyS[1].str.contains('Episode') == False)]

lista_2_peliculas = lista_2_peliculas[(lista_2_peliculas[0].str.contains('Supercampeones') == False) & 
                                      (lista_2_peliculas[0].str.contains('Arcane') == False) &
                                      (lista_2_peliculas[0].str.contains('Viajes Pokémon') == False) &
                                      (lista_2_peliculas[0].str.contains('Clarence') == False) &
                                      (lista_2_peliculas[0].str.contains('DC Super Hero Girls') == False) &
                                      (lista_2_peliculas[0].str.contains('Soy Georgina') == False) &
                                      (lista_2_peliculas[0].str.contains('Kakegurui') == False) &
                                      (lista_2_peliculas[0].str.contains('Woo, una abogada extraordinaria') == False) &
                                      (lista_2_peliculas[0].str.contains('Komi-san no puede comunicarse') == False)]
pd.set_option('display.max_rows', 50)
lista_2_peliculas
```

```python
lista_2_series = lista_2_PyS.drop(lista_2_peliculas.index)
lista_2_series
```

<h3>4.2. Variables derivadas de la fecha</h3>

Extraemos los diferentes componentes de la fecha para generar nuevas variables.

```python
df_netflix['Fecha'] = pd.to_datetime(df_netflix['Date'])
df_netflix.drop(columns = 'Date',inplace = True)
df_netflix
```

```python
df_netflix.set_index('Fecha', inplace = True)
df_netflix
```

Creamos una función para extraer los componentes.

```python
def atributos_fechas(data):
  data['Año'] = data.index.year
  data['Mes'] = data.index.month_name()
  data['Dia_Mes'] = data.index.day
  data['Dia_Semana'] = data.index.day_name()
  return(data)
```

```python
atributos_fechas(df_netflix)
```

<h2>5. Análisis</h2>

¿Desde hace cuánto tiempo tengo contratado Netflix?

```python
from datetime import date

hoy = pd.Timestamp(date.today())
primer_dia = df_netflix.index.min()
tiempo = hoy - primer_dia

print(f'Netflix lleva vivo con nosotros por {tiempo.days} dias.')
```

¿Cuanto me he gastado en Netflix hasta ahora?

```python
coste_mensual = 9

gasto = tiempo.days / 30 * coste_mensual

print(f'Se han gastado {gasto} soles en Netflix.')
```

¿Cuanto tiempo de mi vida le dedico cada año a Netflix?

```python
media_min_serie = 45
media_min_pelicula = 100

consumo = df_netflix.loc[df_netflix['Año'] < 2023].groupby('Tipo').Title.count()

minutos_peliculas_año = consumo['Película'] * media_min_pelicula / 3
minutos_series_año = consumo['Serie'] * media_min_serie / 3

dias_pelis_año = minutos_peliculas_año / 60 / 24
dias_series_año = minutos_series_año / 60 / 24

print(f'Al año se dedican {round(dias_series_año)} dias de tu vida a ver series y {round(dias_pelis_año)} dias de tu vida a ver peliculas')
```
